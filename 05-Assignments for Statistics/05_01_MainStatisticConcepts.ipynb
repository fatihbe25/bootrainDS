```python
import pandas as pd 
import numpy as np 
from scipy import stats

data1=np.append( np.random.normal(30,3,100) ,  np.random.normal(45,8,100))
data2=np.append( np.random.normal(66,15,100) ,  np.random.normal(15,28,100))
data3=np.array([14,24,34,56,74,23,4,5,62,32,33,44,15,45,27,87,31,99,1,22,44,9])

# print("data1:\n",data1)
# print("data2:\n",data2)
# print("data3:\n",data3)

# with formula
mean1=data1.sum()/len(data1)   
mean2=data2.sum()/len(data2)  
mean3=data3.sum()/len(data3)  




data1.sort()
L=len(data1)
if  L%2==0:
    median1=(data1[(L//2)-1] + data1[(L//2)]) / 2 
else:
    median1=data1[(L//2)-1]

data2.sort()
L=len(data2)
if  L%2==0:
    median2=(data2[(L//2)-1] + data2[(L//2)]) / 2 
else:
    median2=data2[(L//2)-1]

data3.sort()
L=len(data3)
if  L%2==0:
    median3=(data3[(L//2)-1] + data3[(L//2)]) / 2 
else:
    median3=data3[(L//2)-1]

    
(values,counts) = np.unique(data1,return_counts=True)
midx=np.argmax(counts)
mode1=values[midx]

(values,counts) = np.unique(data2,return_counts=True)
midx=np.argmax(counts)
mode2=values[midx]

(values,counts) = np.unique(data3,return_counts=True)
midx=np.argmax(counts)
mode3=values[midx]


total=0
for d in data1:
    total+=(d-mean1)**2
variance1=total/(len(data1)-1)
std1=variance1**0.5

total=0
for d in data2:
    total+=(d-mean2)**2
variance2=total/(len(data2)-1)
std2=variance2**0.5

total=0
for d in data3:
    total+=(d-mean3)**2
variance3=total/(len(data3)-1)
std3=variance3**0.5

sem1=std1/((len(data1))**0.5 )
sem2=std2/((len(data2))**0.5 )
sem3=std3/((len(data3))**0.5 )


print("_" * 80)
print("***  With Formula  *** ")
print("_" * 80)
print("# central tendency: mean, median, mode ")
print("mean of data1 : {:.3f}".format(mean1))
print("mean of data2 : {:.3f}".format(mean2))
print("mean of data3 : {:.3f}".format(mean3))
print("median of data1 : {:.3f}".format(median1) )
print("median of data2 : {:.3f}".format(median2) )
print("median of data3 : {:.3f}".format(median3) )
print("mode of data1 : {:.3f}".format(mode1) )
print("mode of data2 : {:.3f}".format(mode2) )
print("mode of data3 : {:.3f}".format(mode3) )
print("# distribution :  variance, standart deviation, standart error")
print("variance of data1 : {:.3f} - standard deviation of data1 : {:.3f}".format(variance1,std1))
print("variance of data2 : {:.3f} - standard deviation of data2 : {:.3f}".format(variance2,std2))
print("variance of data3 : {:.3f} - standard deviation of data3 : {:.3f}".format(variance3,std3))
print("data1 => variance: {:.3f}, standard deviation: {:.3f}, standard error: {:.3f} ".format(variance1,std1,  sem1 ) )
print("data2 => variance: {:.3f}, standard deviation: {:.3f}, standard error: {:.3f} ".format(variance2,std2,  sem2 ) )
print("data3 => variance: {:.3f}, standard deviation: {:.3f}, standard error: {:.3f} ".format(variance3,std3,  sem3 ) )



print("_" * 80)
print("***  With Numpy  *** ")
print("_" * 80)
print("# central tendency: mean, median, mode ")
print("mean of data1 : {:.3f}".format(data1.mean()))
print("mean of data2 : {:.3f}".format(data2.mean()))
print("mean of data3 : {:.3f}".format(data3.mean()))
print("median of data1 : {:.3f}".format(np.median(data1)) )
print("median of data2 : {:.3f}".format(np.median(data2)) )
print("median of data3 : {:.3f}".format(np.median(data3)) )
print("mode of data1 : {:.3f}".format(stats.mode(data1)[0][0] ) ) 
print("mode of data2 : {:.3f}".format(stats.mode(data2)[0][0]) )
print("mode of data3 : {:.3f}".format(stats.mode(data3)[0][0]) )
print("# distribution :  variance, standart deviation, standart error")
print("data1 => variance: {:.3f}, standard deviation: {:.3f}, standard error: {:.3f} ".format(data1.var(ddof=1),data1.std(ddof=1), stats.sem(data1,ddof=1)  ) )
print("data2 => variance: {:.3f}, standard deviation: {:.3f}, standard error: {:.3f} ".format(data2.var(ddof=1), data2.std(ddof=1), stats.sem(data2,ddof=1) ) )
print("data3 => variance: {:.3f}, standard deviation: {:.3f}, standard error: {:.3f} ".format(data3.var(ddof=1), data3.std(ddof=1), stats.sem(data3,ddof=1) ) )
```
