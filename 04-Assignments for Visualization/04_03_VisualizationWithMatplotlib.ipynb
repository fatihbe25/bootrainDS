```python
import pandas as pd 
import matplotlib.pyplot as plt 
from datetime import datetime

startingFrom=datetime.now()- pd.to_timedelta("100day")
#print(startingFrom)

covid_data=pd.read_csv("owid-covid-data.csv",parse_dates=["date"],low_memory=False)
#covid_data.head()
covid_data[(covid_data.location.isin(["Spain","France","Germany","Italy"]) ) 
           & ( covid_data.date>= startingFrom ) ]


fig=plt.figure(figsize=(15,5),dpi=120)

plt.subplot(1,2,1)
plt.bar(covid_data["location"],covid_data["total_cases"],color="red")
```
