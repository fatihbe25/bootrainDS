1. Load the array you saved in the previous lesson from the disk.

```python
import numpy as np 

data1=np.load("data1.npy")

print(data1)
```

2. Display the mean and the standard deviation for each column.

```python
m_c=np.mean(data1,axis=0)
sdv_c=np.std(data1,axis=0)

print("Mead of each column:",m_c)
print("STD Dev of each column:",sdv_c)
```

3. Subtract 1, 25, 25, 10, 4 from columns in order. (Remember it can be subtracted in one line of code.)

```python
dataR=data1-[1,25,25,10]


print("data1:\n",data1)
print("-"*100)
print(" substracted [1,25,10,4]  ==> ")
print("-"*100)
print("dataR:\n",dataR)


```

```python
dataM=data1*2

print("data1:\n",data1)
print("-"*100)
print(" multiplied by 2  ==> ")
print("-"*100)
print("dataM:\n",dataM)
```
