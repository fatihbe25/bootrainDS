````python
import pandas as pd

features=pd.read_csv("features.csv",index_col=0)
target=pd.read_csv("target.csv",index_col=0)


display(features)

display(target)

merge_inner=pd.merge(features,target,on="id",how='inner')
merge_outer=pd.merge(features,target,on="id",how='outer')
merge_left=pd.merge(features,target,on="id",how='left')
merge_right=pd.merge(features,target,on="id",how='right')

#display("inner:",merge_inner)
print("inner:",merge_inner.age.count())
#display("outer:",merge_outer)
print("outer:",merge_outer.age.count())
#display("left:",merge_left)
print("left:",merge_left.age.count())
#display("right:",merge_right)
print("right:",merge_right.age.count())






````
