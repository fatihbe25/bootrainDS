1. Create a DataFrame whose index is the first 10 letters of the alphabet and contains two more columns with first 10 prime numbers and the first 10 fibonacci numbers.

```python
import pandas as pd
import numpy as np 


#function geting first N fibinacci numbers.. 
def fibonacci_list(first_n):
    
    fibonacci=[]
    for n in range(first_n): # list index from 0-9
        item=1
        if n>=2:
            item=fibonacci[n-2] + fibonacci[n-1]

        fibonacci.append(item)
        
    return fibonacci


# function geting first N prime numbers using isPrime function..
def prime_list(first_n):
    prime_numbers=[]
    x=1
    while len(prime_numbers)<first_n:
        if isPrime(x):
            prime_numbers.append(x)
        x+=1;
    return prime_numbers
        
#function checking if a number is prime..        
def isPrime(number):
    if number<2:
        return False
    if number<8:
        if number in [2,3,5,7]:
            return True
        else:
            return False
        
    rem_div_six=number%6
    if rem_div_six not in [1,5]:
        return False
    
    if number%5==0 or number%7==0:
        return False
 
    return True



#data frame creation...:

alf="a,b,c,d,e,f,g,h,i,j".split(',')
print(alf)

df=pd.DataFrame(index=alf)
df.index.name="Letter"

df["Primes"]=prime_list(10)
df["Fibinaccis"]=fibonacci_list(10)

display(df)





```
