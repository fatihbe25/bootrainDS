1. Find the types of variables in the dataset and the missing (null) ratio of each variable.

```python
import pandas as pd 
states=pd.read_csv("data/states_all.csv",index_col="PRIMARY_KEY",low_memory=False)

states.head()
```

```python
states.info()
```

```python
def print_missing_percents():
    missings=(states.isnull().sum()/len(states) ) * 100 
    print("Missing values percents by columns:\n" + "-"*50)
    for i,m in zip(missings.index,missings):
        print("{}  : {:.3f}%".format(i.ljust(30),m))
        
print_missing_percents()
```

2. You may have noticed that our data has a year column. For now, forget about the year data and assume that each observation was made in the same year. Consider how you can fill in missing values for each variable. Filling in the missing values with a value is meaningful for which variables and for which ones?

```python
Tree types of variables exists in dataset; the revenue, expenditure and grades values. 
To fill the missing value without affecting the our target result, we can safely fill the missings values 
by the average of that variable for whole dataset.
```

```python
# for each column having missing value, filter missing value containing cells and assing average of the field to that cells.

for column,mis_percent in zip(missings.index,missings):
    if mis_percent==0:
        continue
    #print("filling na for {} with {}".format(column,states[column].mean()))
    states[column].fillna(states[column].mean(),inplace=True)

print_missing_percents()
```

3. Now it's time to consider the factor of time! Review your answer in question 2 and fill in the missing data based on the values observed during that year. For example, if you want to fill a value with an average value, calculate the average of that year.

```python
#reload data again.

states=pd.read_csv("data/states_all.csv",index_col="PRIMARY_KEY",low_memory=False)
states.head()
```

```python
# for each column having missing value, filter missing value containing cells and assing average of the field for that year to that cells.

for column in states.columns:
    if states[column].isnull().sum()>0:
        print("filling na for " + column)
        if states[column].isnull(): 
            states[column].fillna(states.groupby("YEAR")[column].transform('mean'), inplace=True )
        print(states.groupby(["YEAR"])[column].mean())

print_missing_percents()
```
